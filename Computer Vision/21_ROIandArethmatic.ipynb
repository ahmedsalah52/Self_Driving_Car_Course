```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('Files/carkyo.jpg')
```

```python
plt.imshow(img[:,:,::-1]);
```

```python
logo = cv2.imread('Files/logo.png')
```

```python
plt.imshow(logo[:,:,::-1]);
```

```python

print("image size :",img.shape)
print("logo size :" , logo.shape)
```

```python
logo = cv2.resize(logo, (150, 30))
plt.imshow(logo[:,:,::-1]);
```

```python
print("image size :",img.shape)
print("logo size :" , logo.shape)
```

```python
gray = cv2.cvtColor(logo, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray');
```

```python
ret, mask = cv2.threshold(gray, 50, 255, cv2.THRESH_BINARY_INV)
plt.imshow(mask, cmap='gray');
```

```python
mask_RGB = cv2.cvtColor(mask,cv2.COLOR_GRAY2RGB)
plt.imshow(mask_RGB);
```

```python
img_with_logo = img.copy()
img_with_logo[728:758, 1205:1355] &= mask_RGB
```

```python
plt.imshow(img_with_logo[:,:,::-1]);
```

```python
img_with_logo[728:758, 1205:1355] += logo
plt.imshow(img_with_logo[:,:,::-1]);
```

```python
cv2.imwrite('Files/lab2_output.png', img_with_logo)
```
