```python
#original gray image
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread('Files/001.jpg',0)
plt.imshow(img,cmap = 'gray')
```

```python
plt.hist(img.flatten(),256,[0,256]); plt.show()
```

```python
#gray with general equalization
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
equ = cv.equalizeHist(img)
plt.imshow(equ,cmap = 'gray')
```

```python

plt.hist(equ.flatten(),256,[0,256]); plt.show()
```

```python
#gray adabtive equalization

import numpy as np
import cv2 as cv
# create a CLAHE object (Arguments are optional).
clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
cl1 = clahe.apply(img)
plt.imshow(cl1,cmap = 'gray')
```

```python
plt.hist(cl1.flatten(),256,[0,256]); plt.show()
```

```python
#original rgb image

import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread('Files/001.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
#rgb with general equalization

import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
r, g, b = cv.split(img)
r_ = cv.equalizeHist(r)
g_ = cv.equalizeHist(g)
b_ = cv.equalizeHist(b)

rgb = cv.merge([r_, g_, b_])

plt.imshow(rgb)
```

```python
#rgb adabtive equalization

import numpy as np
import cv2 as cv
img = cv.imread('Files/001.jpg')
# create a CLAHE object (Arguments are optional).
clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
r, g, b = cv.split(img)
r_ = clahe.apply(r)
g_ = clahe.apply(g)
b_ = clahe.apply(b)

rgb = cv.merge([r_, g_, b_])
plt.imshow(rgb,cmap = 'gray')
```
