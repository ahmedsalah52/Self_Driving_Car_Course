```python
import cv2

image = np.zeros((500, 500), np.uint8)

def nothing(x):
    pass

cv2.namedWindow('Image')

cv2.createTrackbar('row', 'Image', 0, image.shape[0]-1, nothing)
cv2.createTrackbar('col', 'Image', 0, image.shape[1]-1, nothing)
cv2.createTrackbar('x', 'Image', 0, image.shape[0]-1, nothing)
cv2.createTrackbar('y', 'Image', 0, image.shape[1]-1, nothing)

while 1:
    row = cv2.getTrackbarPos('row', 'Image')
    col = cv2.getTrackbarPos('col', 'Image')
    x = cv2.getTrackbarPos('x', 'Image')
    y = cv2.getTrackbarPos('y', 'Image')
    
    image = np.zeros((500, 500), np.uint8)
    
    image[row, :] = 255
    image[:, col] = 255

    image[x, x] = 255

    cv2.imshow('Image', image)
    k = cv2.waitKey(1)
    if k == 27:
        break

cv2.destroyAllWindows()
```
